```python
import time
import pyautogui 
import tkinter as tk
from tkinter import messagebox
pyautogui.PAUSE = 0.5
import cv2
import numpy as np
import locale
import subprocess


# definição de configuração de imagens 
def locate_image_opencv(image_path, confidence=0.8):
    try:
        # Tira uma captura de tela
        screenshot = pyautogui.screenshot()
        screenshot = cv2.cvtColor(np.array(screenshot), cv2.COLOR_RGB2BGR)
        
        # Carrega a imagem do template
        template = cv2.imread(image_path, cv2.IMREAD_COLOR)
        if template is None:
            print(f"Erro: A imagem do template em {image_path} não pôde ser carregada.")
            return None
        
        # Realiza a correspondência do template
        result = cv2.matchTemplate(screenshot, template, cv2.TM_CCOEFF_NORMED)
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
        
        if max_val >= confidence:
            return max_loc
        return None

    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        return None
    
def localizar_imagem_opencv(img_path, confidence=0.8):
    screenshot = pyautogui.screenshot()
    screenshot = cv2.cvtColor(np.array(screenshot), cv2.COLOR_RGB2BGR)
    template = cv2.imread(img_path, cv2.IMREAD_COLOR)
    res = cv2.matchTemplate(screenshot, template, cv2.TM_CCOEFF_NORMED)
    loc = np.where(res >= confidence)
    if len(loc[0]) > 0 and len(loc[1]) > 0:
        return loc[1][0], loc[0][0]
    return None    
#-----------------------------------------------------------------------------------------------

# definição para localizar imagem clicando com o botão direito do mouse 
def localiza_rigth(img):

    while True:
        location = locate_image_opencv(img, confidence=0.8)
        if location:
            template = cv2.imread(img, cv2.IMREAD_COLOR)
            h, w, _ = template.shape
            center_x = location[0] + w // 2
            center_y = location[1] + h // 2
            pyautogui.click(center_x, center_y, button='right')
            print("Imagem encontrada e clicada!")
            break  # Sai do loop após encontrar e clicar na imagem
        else:
            print("Imagem não encontrada. Tentando novamente...")
        
        time.sleep(1)  # Aguarda 1 segundo antes de tentar novamente
#------------------------------------------------------------------------------------------------
# definição para localizar imagens 
def localiza(img):
    while True:
        location = locate_image_opencv(img, confidence=0.8)
        if location:
            template = cv2.imread(img, cv2.IMREAD_COLOR)
            h, w, _ = template.shape
            center_x = location[0] + w // 2
            center_y = location[1] + h // 2
            pyautogui.click(center_x, center_y,)
            print("Imagem encontrada e clicada!")
            return True  # Sai do loop após encontrar e clicar na imagem
        else:
            print("Imagem não encontrada. Tentando novamente...")
        
        time.sleep(1)  # Aguarda 1 segundo antes de tentar novamente
        
#------------------------------------------------------------------------------------------------
#definição para localizar imagens com 2 clicks
def localiza2_clicks(img):
    while True:
        location = locate_image_opencv(img, confidence=0.8)
        if location:
            template = cv2.imread(img, cv2.IMREAD_COLOR)
            h, w, _ = template.shape
            center_x = location[0] + w // 2
            center_y = location[1] + h // 2
            pyautogui.click(center_x, center_y, clicks=2)
            print("Imagem encontrada e clicada!")
            break  # Sai do loop após encontrar e clicar na imagem
        else:
            print("Imagem não encontrada. Tentando novamente...")
        
        time.sleep(1)  # Aguarda 1 segundo antes de tentar novamente

#-----------------------------------------------------------------------------------------------
# definição para localizar quando a imagem desaparecer 
def localiza_imgsome(img):
    template = cv2.imread(img, cv2.IMREAD_COLOR)
    h, w, _ = template.shape
    img_presente = True

    while img_presente:
        location = localizar_imagem_opencv(img, confidence=0.8)
        print(location)
        if location is not None:
            print("Sendo gerada...")
            time.sleep(1)
            img_presente = True
            # Aqui você coloca o código que deseja executar quando a imagem desaparecer            
        else:
            pyautogui.press("esc")
            print("Gerada")
            img_presente = False

#----------------------------------------------------------------------------------------------- 

# Definição para localizar quando a imagem aparecer, com tempo limite de 3 segundos
def localiza_alerta(img):
    template = cv2.imread(img, cv2.IMREAD_COLOR)
    h, w, _ = template.shape
    img_presente = False
    tempo_inicio = time.time()  # Marca o tempo inicial

    while not img_presente:
        # Verifica se o tempo limite de 3 segundos foi atingido
        tempo_atual = time.time()
        if tempo_atual - tempo_inicio > 3:
            print("Tempo limite de 3 segundos atingido. A imagem não foi encontrada.")
            break

        location = localizar_imagem_opencv(img, confidence=0.8)
        print(location)
        if location is not None:
            localiza(botao_alerta)
            print("Imagem apareceu, 'Enter' pressionado.")
            img_presente = True  # Sai do loop ao encontrar a imagem
        else:
            print("Imagem ainda não apareceu...")
            time.sleep(0.5)  # Espera 0,5 segundos antes de verificar novamente
#------------------------------------------------------------------------------------------------            
# aparece um popup quando estiver atualizando o shell 

def show_popup_2(message):
    root = tk.Tk()
    root.withdraw()  # Oculta a janela principal
    messagebox.showinfo("Informação", message)
    root.destroy()


#-----------------------------------------------------------------------------------------------


usuario = 'algarcia'
senha = 'Atacadao@071106'

# imagens localizadas
putty = r'C:\Users\algarcia\Documents\comands\img\putty.png'
db739 = r'C:\Users\algarcia\Documents\comands\img\db739.png'
load = r'C:\Users\algarcia\Documents\comands\img\load.png'
abrir = r'C:\Users\algarcia\Documents\comands\img\abrir.png'
maximiza = r'C:\Users\algarcia\Documents\comands\img\maximiza.png'
alerta = r'C:\Users\algarcia\Documents\comands\img\alerta.png'
botao_alerta = r'C:\Users\algarcia\Documents\comands\img\botao_alerta.png'
```
